# **Introduction**
This case study delves into a leading Indian private sector bank's initiative to optimize its personal loan campaigns through sophisticated data exploration and predictive modeling techniques. By leveraging advanced analytics, the bank aims to better understand customer behavior and preferences, ultimately improving the effectiveness of its marketing strategies.

# **Business Challenge**
**1.** The bank faces the critical challenge of identifying potential customers within its extensive customer base who are most likely to accept personal loan offers.

**2.** This requires a deep analysis of various factors that influence a customer's decision to take a personal loan.

**3.** The goal is to develop a data-driven strategy that enables the bank to target these potential customers more effectively, thereby increasing the success rate of its marketing campaigns and enhancing overall customer satisfaction.

# **About the Dataset**

**1. ID:** Customer ID

**2. Pin-code:** Home Area pincode

**3. Age:** Customer's age

**4. Fam Members:** Total Members in the Family

**5. Education:** Customer's Education

**6. T-Experience:** Years Of Professional Experience

**7. Income:** Annual Income of the Customer

**8. Mortgage:** Value Of the House Mortgage(if any)

**9. Fixed Deposit:** Does the customer have a certificate of deposit.(CD) account with the bank?

**10. Demat:** Does the customer have a demat account with the bank?

**11. Net Banking:** Does the customer use internet banking facilities.

**12. Loan:** Did this customer accept the personal loan offered in the last campaign?


# **Libraries used in the projects**

### 1. **`numpy`**:
   - **Purpose**: `numpy` is a fundamental library for numerical computing in Python. It provides support for large multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays.
   - **Use Cases**: Handling arrays, performing mathematical operations, and working with linear algebra.

### 2. **`pandas`**:
   - **Purpose**: `pandas` is a powerful data manipulation and analysis library. It provides data structures like Series and DataFrames, which are ideal for handling structured data.

### 3. **`matplotlib.pyplot`**:
   - **Purpose**: `matplotlib.pyplot` is a plotting library used for creating static, animated, and interactive visualizations in Python.
   - **Use Cases**: Creating plots, charts, and graphs (e.g., line plots, bar charts, scatter plots).

### 4. **`seaborn`**:
   - **Purpose**: `seaborn` is a data visualization library based on `matplotlib`. It provides a high-level interface for drawing attractive and informative statistical graphics.
   - **Use Cases**: Enhancing visualizations with themes, adding statistical plots like box plots, violin plots, and pair plots.

### 5. **`sklearn.model_selection.train_test_split`**:
   - **Purpose**: This function from `scikit-learn` is used to split your dataset into training and testing sets.
   - **Use Cases**: Splitting data for training machine learning models and evaluating their performance.

### 6. **`sklearn.linear_model.LinearRegression`**:
   - **Purpose**: This is a linear regression model provided by `scikit-learn`.
   - **Use Cases**: Predicting a continuous target variable based on one or more input features.

### 7. **`sklearn.metrics.mean_squared_error`, `r2_score`**:
   - **Purpose**: These functions from `scikit-learn` are used to evaluate the performance of regression models.
   - **Use Cases**:
     - `mean_squared_error`: Measures the average of the squares of the errors (the difference between the predicted and actual values).
     - `r2_score`: Represents the proportion of the variance in the dependent variable that is predictable from the independent variables.

### 8. **`sklearn.preprocessing.StandardScaler`, `MinMaxScaler`**:
   - **Purpose**: These are preprocessing tools from `scikit-learn` used to scale features.
   - **Use Cases**:
     - `StandardScaler`: Scales features to have a mean of 0 and a standard deviation of 1.
     - `MinMaxScaler`: Scales features to a given range, typically [0, 1].

### 9. **`sklearn.decomposition.PCA`**:
   - **Purpose**: PCA (Principal Component Analysis) is a technique from `scikit-learn` used for dimensionality reduction.
   - **Use Cases**: Reducing the number of features in a dataset while preserving as much variance as possible.

### 10. **`tensorflow` and `keras`**:
   - **Purpose**: `tensorflow` is an open-source machine learning framework, and `keras` is a high-level API for building and training deep learning models. `Keras` is integrated into TensorFlow as `tf.keras`.

### 11. **`tensorflow.keras.models.Sequential`**:
   - **Purpose**: This is a Keras model type that allows you to build a linear stack of layers for a neural network.
   - **Use Cases**: Creating simple feed-forward neural networks by stacking layers.

### 12. **`tensorflow.keras.layers.Dense`, `Dropout`**:
   - **Purpose**: These are Keras layers used to build neural networks.
   - **Use Cases**:
     - `Dense`: A fully connected layer where each input is connected to each output.
     - `Dropout`: A regularization layer that randomly sets a fraction of input units to 0 at each update during training, which helps prevent overfitting.

### 13. **`scipy.stats`**:
   - **Purpose**: `scipy.stats` is a module in SciPy that contains a large number of probability distributions and statistical functions.
   - **Use Cases**: Performing statistical tests, calculating distributions, and working with random variables.

### 14. **`mplcyberpunk`**:
